```python
import sqlite3

conn = sqlite3.connect('silq.db')
cursor = conn.cursor()
```

```python
query = "SELECT name FROM sqlite_master WHERE type='table';"

# Execute the query
cursor.execute(query)

# Fetch all the results
tables = cursor.fetchall()

# Extract table names from the results
table_names = [table[0] for table in tables]

# Print the table names
for name in table_names:
    print(name)
```

```python
from sqlalchemy import create_engine

engine = create_engine('sqlite:///silq.db')
```

```python
engine
```

```python
sql = """
WITH ords AS (
    SELECT 
        ProductID, 
        COUNT(OrderID) AS order_cnt 
    FROM 
        dm_order 
    GROUP BY 
        ProductID
) 
SELECT 
    dm_product.ProductID, 
    dm_product.ProductName, 
    COALESCE(ords.order_cnt, 0) AS order_cnt 
FROM 
    dm_product 
LEFT JOIN 
    ords ON dm_product.ProductID = ords.ProductID 
ORDER BY 
    order_cnt
"""
```

```python
analysis = pd.read_sql_query(sql, conn)
```

```python
analysis
```

```python
sql2 = """
WITH ords AS (
SELECT 
    dm_order.UserID,
    dm_order.ProductID,
    dm_product.Price,
    (Amount - Price) as Profit
FROM 
    dm_order 
LEFT JOIN 
    dm_product ON dm_product.ProductID = dm_order.ProductID
)
select
    dm_user.UserID,
    coalesce(sum(ords.Profit),0) as TotalProfit
from dm_user 
left join
    ords on dm_user.UserID = ords.UserID
group by dm_user.UserID
order by TotalProfit desc
"""
```

```python
analysis2 = pd.read_sql_query(sql2, conn)
```

```python
analysis2
```

```python
sql3 = """
select
    dm_user.UserID,
    coalesce(sum(ords.Amount),0) as TotalRevenue
from dm_user 
left join
    dm_order ords on dm_user.UserID = ords.UserID
group by dm_user.UserID
order by TotalRevenue desc
"""
```

```python
analysis3 = pd.read_sql_query(sql3, conn)
```

```python
analysis3
```

```python
sql4 = """
SELECT 
    strftime('%Y-%m', SignUpDate) AS CohortMonth,
    COUNT(UserID) AS NumberOfUsers
FROM 
    dm_user
GROUP BY 
    CohortMonth
ORDER BY 
    CohortMonth;
"""
analysis4 = pd.read_sql_query(sql4, conn)
```

```python
analysis4
```

```python
sql5 = """
WITH MonthlyCohorts AS (
    SELECT
        UserID,
        strftime('%Y-%m', SignUpDate) AS Cohort
    FROM dm_user
),
UserActivity AS (
    SELECT
        o.UserID,
        c.Cohort,
        strftime('%Y-%m', o.OrderDate) AS ActivityMonth
    FROM MonthlyCohorts c
    JOIN dm_order o ON c.UserID = o.UserID
)
select * from UserActivity 

"""
analysis5 = pd.read_sql_query(sql5, conn)
analysis5
```
