### Q1. List the total number of orders for each product, sorted by the number of orders.

```python
sql = """
WITH ords AS (
    SELECT 
        ProductID, 
        COUNT(OrderID) AS order_cnt 
    FROM 
        Orders 
    GROUP BY 
        ProductID
) 
SELECT 
    Products.ProductID, 
    Products.ProductName, 
    COALESCE(ords.order_cnt, 0) AS order_cnt 
FROM 
    Products 
LEFT JOIN 
    ords ON Products.ProductID = ords.ProductID 
ORDER BY 
    order_cnt DESC
"""
```

```python
analysis = pd.read_sql_query(sql, conn)
```

```python
analysis
```

### Q2. Find the total revenue generated by each user.

```python
sql3 = """
select
    Users.UserID,
    coalesce(sum(ords.Amount),0) as TotalRevenue
from Users 
left join
    Orders ords on Users.UserID = ords.UserID
group by Users.UserID
order by TotalRevenue desc
"""
```

```python
analysis3 = pd.read_sql_query(sql3, conn)
```

```python
analysis3
```

### Q3. Display users who have not placed any orders.

```python
sql4 = """
with main as (
select
    Users.UserID,
    coalesce(sum(ords.Amount),0) as TotalRevenue
from Users 
left join
    Orders ords on Users.UserID = ords.UserID
group by Users.UserID
order by TotalRevenue desc
)
select * from main where TotalRevenue = 0
"""
```

```python
analysis4 = pd.read_sql_query(sql4, conn)
```
